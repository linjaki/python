```python
使用 import requests 可以讓你在 Python 程式中方便地發送 HTTP 請求（如 GET、POST 等），常用於存取 API 或抓取網頁資料。以下是詳細的使用方法：

1. 安裝 requests 套件 pip install requests
2. 匯入模組 在 Python 檔案開頭加上： import requests
3. 基本用法 
# 發送 GET 請求 
    response = requests.get('https://www.example.com')
    print(response.status_code)  # 顯示 HTTP 狀態碼
    print(response.text)         # 顯示網頁內容
# 發送帶參數的 GET 請求
params = {'key1': 'value1', 'key2': 'value2'}
response = requests.get('https://www.example.com', params=params)
print(response.url)  # 查看實際請求的網址
```

```python
import requests
from requests import Response


url:str = 'https://raw.githubusercontent.com/linjaki/python/refs/heads/main/lesson10/students.csv'

# 1. 發送 GET 請求並取得網頁內容
response:Response = requests.get(url)

## 印出狀態碼，200 表示成功
if response.status_code == 200:
    print("下載成功")
    content:str = response.text
```

```python
print(content)
```

## import io 
是 Python 內建的標準函式庫，用於處理各種檔案和資料流（stream）的操作，
特別適合處理文字和二進位資料的讀寫。這個模組常用於記憶體中的檔案操作，
例如：將字串當作檔案來讀寫、處理二進位資料等。

```python
import io
from io import StringIO
import csv

file:StringIO = io.StringIO(content)
reader = csv.DictReader(file)
students = [row for row in reader]
file.close()
students
```
