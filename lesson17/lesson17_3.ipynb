```python
import numpy as np
import pandas as pd

np.random.seed(0)
df = pd.DataFrame(np.random.randn(10,4), columns=['A', 'B', 'C', 'D'])
df
```

```python
df.iloc[0,2] = np.nan
df.iloc[4, 3] = np.nan
df
```

```python
df.style.highlight_null(color='red')
```

# df.loc[:4]
取出 index 0~4 的前 5 列資料。
# .style.highlight_max(axis=1, ...)
針對每一列（axis=1），將最大值的儲存格套用指定樣式。
# props="color:while; font-weight: bold; background-color:darkblue"
指定最大值的樣式：
# 字體顏色（color）應為 white、字體加粗（font-weight: bold）、背景色為深藍色（background-color:darkblue）

```python
df.loc[:4].style.highlight_max(axis=1,props="color:white; font-weight: bold; background-color:darkblue")
```

```python

df.style.highlight_between(left=[1.4, 2.4, 3.4], right=[1.6, 2.6, 3.6], axis=1, props="color:white; font-weight:bold; background-color:darkblue")
```
