## numoy & pandas 區別


```python
import numpy as np
import pandas as pd

np.random.randint(50, 101, size=(50, 5))
```

# 會產生一個 50×5 的二維 NumPy 陣列，元素是隨機整數，範圍從 50（含）到 100（含），不包含 101。
這通常用來模擬 50 位學生在 5 個科目上的分數，或是其他需要大量隨機整數資料的場合。
50：最小值（包含）。
101：最大值（不包含，所以實際最大值是 100）。
size=(50, 5)：產生 50 列 5 欄的二維陣列。
-這種隨機資料生成方式非常適合用於數值運算、統計分析或資料科學的練習

```python
scores = np.random.randint(50, 101, size=(50, 5))
```

利用 NumPy 產生一個 50×5 的二維整數陣列 scores，每個元素都是隨機產生的分數，
範圍在 50 到 100（包含 50，不包含 101，即 50～100 之間的整數）
-模擬大量數據、隨機測試資料、統計分析等數值運算場景

```python
df = pd.DataFrame(scores,
             columns = ["國文","英文","地理","人文","數學"],
             index=[f"stu{i}" for i in range(1,51)])
df.loc['stu1'].name
```

# 將前面用 NumPy 產生的隨機分數陣列 scores 轉換成一個有標題和索引的 pandas DataFrame，方便進行資料分析和操作。
pd.DataFrame(scores, ...) 把 scores 這個 50×5 的二維陣列轉成 DataFrame。
columns=["國文", "英文", "地理", "人文", "數學"] 指定每一欄的名稱，分別對應五個科目。
index=[f"student{i}" for i in range(1,51)] 指定每一列的名稱，從 student1 到 student50，對應 50 位學生。
df 這個 DataFrame 會像一個表格，列是學生，欄是科目，內容是分數。
-DataFrame 結構非常適合進行統計、資料分析、視覺化等後續操作，特別是在數值運算和資料處理場景中

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(scores,
             columns = ["國文","英文","地理","人文","數學"],
             index=[f"stu{i}" for i in range(1,51)])
name1 = df.loc['stu1'].name
name2 = df.loc['stu2'].name       
figure = plt.figure(figsize=(6, 5),dpi=150,facecolor='#eeeeee')
axes = figure.add_subplot(1,1,1)
axes.plot(df.columns,df.loc['stu1'],marker='o',label=name1)
axes.plot(df.columns,df.loc['stu2'],marker='o',linestyle='-.',label=name2)
axes.set_title(f"{name1} vs {name2}")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.legend()
xlocations = axes.get_xticks()
for index in range(len(xlocations)):
    x = xlocations[index]
    student1_y = df.loc['stu1'][index]
    student2_y = df.loc['stu2'][index]
    # print("x座標:",x)
    # print("s1_y_座標:",student1_y)
    # print("s2_y_座標:",student2_y)
    # print("==================")
    axes.annotate(str(student1_y),(x-0.05,student1_y+1))
    if student2_y == 98:
        axes.annotate(str(student2_y),(x,student2_y),xytext=(x+0.3, student2_y+5), arrowprops=dict(facecolor='red', shrink=0.05))
    else:
        axes.annotate(str(student2_y),(x-0.05,student2_y+1))
    axes.set_xlim(-0.5,4.5)
    axes.set_ylim(50,105)
axes.grid(axis='y',color='blue',alpha=0.1,linestyle=':')
plt.show()
```
