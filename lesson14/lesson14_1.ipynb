```python
import math
import matplotlib.pyplot as plt

t = np.arange(0,2.5,0.1)
y1 = np.sin(t * math.pi)
y2 = np.sin(t * math.pi+ math.pi/2)
y3 = np.sin(t * math.pi- math.pi/2)
plt.plot(t, y1, marker="*", color="blue", linestyle=" ")
plt.plot(t, y2, marker="^", color="green", linestyle="--")
plt.plot(t, y3,marker="s", color='yellow',linestyle='-')
plt.savefig('a.png')
plt.show()
plt.close()
```

```python
import math 匯入 math 模組（用於數學常數與函數，如 math.pi）
import matplotlib.pyplot as plt 匯入 matplotlib.pyplot，這是常用的 Python 資料視覺化繪圖工具
```

```python
t = np.arange(0,2.5,0.1) 產生一個從 0 到 2.4（不含 2.5），間隔為 0.1 的數值序列，作為 x 軸（時間軸）
y1 = np.sin(t * math.pi) y1：標準正弦波，輸入為t×π
y2 = np.sin(t * math.pi+ math.pi/2) y2：將正弦波相位向左移 π/2 （+90度），波形提前。
y3 = np.sin(t * math.pi- math.pi/2) 將正弦波相位向右移 π/2 （-90度），波形延後。
```

```python
plt.plot(t, y1, marker="*", color="blue", linestyle=" ") 第一條（y1）：藍色星號標記，無連線（linestyle=" "）。
plt.plot(t, y2, marker="^", color="green", linestyle="--") 第二條（y2）：綠色三角標記，虛線連線。
plt.plot(t, y3, marker="s", color='yellow', linestyle='-') 第三條（y3）：黃色方塊標記，實線連線。
```

```python
plt.savefig('a.png') 儲存圖形到本地檔案 'a.png'。
plt.show() 顯示圖形視窗。
plt.close() 關閉圖形，釋放記憶體資源。
```
