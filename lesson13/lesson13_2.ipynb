```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, marker='o',label='line plot',linestyle='-.',color='r',markersize=10,linewidth=2)
axes.set_title("Line Plot on Existing Axes")
axes.legend()
plt.show()
```

```python
# 使用 Python 的 Matplotlib 函式庫來繪製一個折線圖
import matplotlib.pyplot as plt
-匯入 Matplotlib 中的 pyplot 模組，常用來繪製各種圖表
figure, axes = plt.subplots()
-建立一個圖形（figure）和一組座標軸（axes）。plt.subplots() 預設產生一個圖和一個座標軸物件，方便後續繪圖
x = [0, 1, 2]
y = [3, 4, 5]
-定義 x 軸和 y 軸的資料點，這裡 x 軸有三個點 0、1、2，y 軸對應的點為 3、4、5
```

```python
# 座標軸 axes 上繪製折線圖：
-x, y：指定繪圖的座標點。
-marker='o'：每個資料點以圓圈標記。
-label='line plot'：設定圖例中該線的標籤名稱。
-linestyle='-.'：折線的樣式為點劃線（dash-dot line）。
-color='r'：線條顏色為紅色。
-markersize=10：標記圓圈的大小為 10。
-linewidth=2：線條寬度為 2
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mpl

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font',family='ChineseFont')

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, marker='o',label='學生1',linestyle='-.',color='r',markersize=10,linewidth=2)
axes.set_title("折線圖")
axes.set_xticks([0,1,2]),['0','1','2']
axes.set_yticks([3,4,5]),['3','4','5']
axes.legend()
plt.show()
```
