# 利用groupby建立索引做成樞紐分析
groupby機制 (時間區段、數值範圍)
切割-統計

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({"key1":['a','a','b','b','a'],
                     "key2":['one','two','one','two','one'],
                     "data1": np.random.randn(5),
                     "data2":np.random.randn(5)})
df
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                  'key2':['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
df
```

```python

df.groupby(by="key1")[['data1','data2']].sum()
```

```python
df1 = df.groupby(by=['key1','key2']).sum()
df1.columns.name = "data"
df1
```

```python
df1.stack().unstack(level=['key1','key2']).sum()
df1.columns.name = "data"
df1
```

```python
df1.stack().unstack(level=['key2','data'])
```

groupby物件-->統計的meth

```python
df1 = df.groupby(by=['key1','key2']).agg([("加總","sum"),("平均","mean")])
df1
```
